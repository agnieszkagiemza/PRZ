```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from collections import Counter

# Zakładam, że kolumna z frazami nazywa się 'text'
phrases = df['Category'].astype(str).tolist()

# Zliczam wystąpienia poszczególnych fraz
phrase_counts = Counter(phrases)

# Tworzenie chmury słów z frazami bez przetwarzania
wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(phrase_counts)


# Rysowanie chmury słów
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')  # Ukrycie osi
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Przykładowe dane (zastąp je swoimi danymi)
# df = pd.DataFrame({'App': ['App1', 'App2', 'App3', 'App4'], 'Installs': [100, 200, 300, 400]})

# Wykres słupkowy z 'App' na osi X i 'Installs' na osi Y
plt.figure(figsize=(12, 6))
sns.barplot(x='App', y='Installs', data=df, palette='viridis')
plt.xlabel('Nazwy Aplikacji')
plt.ylabel('Liczba Instalacji')
plt.title('Instalacje Aplikacji')
plt.xticks(rotation=45)  # Rotacja etykiet na osi X dla lepszej czytelności
plt.tight_layout()  # Ułatwia dopasowanie elementów wykresu
plt.show()
```
